```python
cc_df.drop(['merchant','first','last','street','city','state','zip','job','trans_num','unix_time'], axis = 1, inplace= True)        # 불필요한 컬럼들을 제거
```

```python
# 구매 금액(amt) 관련 feature 분석
```

```python
# cc_num(신용카드 번호)기준으로 구매금액의 z-score 계산하기
# 효과: 사용자별 평소 소비 패턴을 벗어난 거래 금액 탐지에 유용합니다.
```

```python
cc_df['cc_num'].value_counts()      # cc_num 컬럼의 값마다 데이터 개수 확인
```

```python
amt_info = cc_df.groupby('cc_num')['amt'].agg(['mean','std']).reset_index()     # cc_num 컬럼의 값마다 amt의 평균, 표준편차를 계산하여 amt_info에 저장
```

```python
amt_info.to_pickle('./amt_info.pkl')        # amt_info라는 DataFrame을 피클(pickle) 형식으로 저장
```

```python
cc_df = cc_df.merge(amt_info, on = 'cc_num', how = 'left')      # cc_num 컬럼을 기준으로, cc_df와 amt_info 데이터를 합쳐서 cc_df에 저장
```

```python
cc_df['amt_z'] = (cc_df['amt'] - cc_df['mean']) / cc_df['std']      # 결제금액(amt)의 z-score(표준점수)를 계산하여 amt_z 컬럼에 저장
```

```python
cc_df[cc_df['is_fraud'] == 1]       # 사기 거래 데이터(is_fraud=1) 확인
```

```python
cc_df.drop(['mean','std'], axis = 1, inplace = True)        # z-score 계산이 완료되었으니 mean, std 컬럼 제거
```

```python
# 구매 금액(amt) 관련 feature 분석
```

```python
# cc_num(신용카드 번호)와 category(구매 상점 분류)를 기준으로 구매금액의 z-score 계산하기
# 효과: 개인별, 상점 분류별 이상 거래 금액 탐지에 유용합니다. 즉, 상황을 고려한 이상 탐지를 가능하게 합니다.
```

```python
cat_info = cc_df.groupby(['cc_num','category'])['amt'].agg(['mean','std']).reset_index()        # cc_num, category 컬럼의 값마다 결제금액(amt)의 평균, 표준편차를 계산하여 cat_info에 저장
```

```python
cat_info.to_pickle('./cat_info.pkl')
```

```python
cc_df = cc_df.merge(cat_info, on = ['cc_num','category'], how = 'left')     # cc_num 컬럼을 기준으로, cc_df와 cat_info 데이터를 합쳐서 cc_df에 저장
```

```python
cc_df['cat_amt_z'] = (cc_df['amt'] - cc_df['mean']) / cc_df['std']      # 결제금액(amt)의 z-score를 계산하여 cat_amt_z 컬럼에 저장
```

```python
cc_df.drop(['mean','std'], axis =1 , inplace = True)        # z-score 계산이 완료되었으니 mean, std 컬럼 제거
```

```python
cc_df.head()
```

```python
# 결제 시간 관련 feature 분석
```

```python
# 한 카드가 특정 시간대에 결제할 확률 또는 비중 계산하기
# 효과: 카드 사용자의 평소 시간대별 소비 패턴에서 벗어난 거래를 감지하는 근거로 사용됩니다.
```

```python
cc_df['hour'] = pd.to_datetime(cc_df['trans_date_trans_time']).dt.hour      # trans_date_trans_time 컬럼에서 시간 값을 추출하여 hour 컬럼에 저장
```

```python
cc_df.head()
```

```python
def hour_func(x):           # 결제시간을 morning, afternoon, night, evening으로 분류하기 위해 함수를 정의
    if (x >= 6) & (x < 12):
        return 'morning'
    elif (x >= 12) & (x < 18):
        return 'afternoon'
    elif (x >= 18) & (x < 23):
        return 'night'
    else:
        return 'evening'
```

```python
cc_df['hour_cat'] = cc_df['hour'].apply(hour_func)      # hour 컬럼에 hour_func 함수를 적용한 값을 hour_cat 컬럼에 저장
```

```python
cc_df.head()
```

```python
cc_df['hour_cat'].value_counts      # 시간대별로 데이터 수를 계산하여 출력
```

```python
all_cnt = cc_df.groupby('cc_num')['amt'].count().reset_index()      # cc_num 컬럼의 값마다 amt 컬럼의 데이터 수를 계산하여 all_cnt 변수로 저장
```

```python
hour_cnt = cc_df.groupby(['cc_num','hour_cat'])['amt'].count().reset_index()        # cc_num, hour_cat 컬럼의 값마다 amt 컬럼의 데이터 수를 계산하여 hour_cnt 변수로 저장
```

```python
all_cnt.head()
```

```python
hour_cnt.head()
```

```python
hour_cnt = hour_cnt.merge(all_cnt, on = 'cc_num', how = 'left')     # cc_num 컬럼을 기준으로 hour_cnt와 all_cnt를 병합
```

```python
hour_cnt.head()
```

```python
hour_cnt = hour_cnt.rename({'amt_x': 'hour_cnt', 'amt_y': 'total_cnt'}, axis = 1)       # amt_x와 amt_y 컬럼의 이름을 각각 hour_cnt, total_cnt로 변경
```

```python
hour_cnt.head()
```

```python
hour_cnt['hour_perc'] = hour_cnt['hour_cnt'] / hour_cnt['total_cnt']        # 전체 거래 건수(total_cnt)에 대한 시간대별 거래 건수(hour_cnt)의 비율을 계산하여 hour_perc 컬럼에 저장
```

```python
hour_cnt.head(10)
```

```python
hour_cnt.tail(10)
```

```python
cc_df.head()
```

```python
hour_cnt = hour_cnt[['cc_num','hour_cat','hour_perc']]      # cc_num, hour_cat, hour_perc 세 컬럼만 남겨서 hour_cnt 변수에 저장
```

```python
hour_cnt.to_pickle('./hour_cnt.pkl')
```

```python
cc_df = cc_df.merge(hour_cnt, on = ['cc_num','hour_cat'], how = 'left')     # cc_num, hour_cat 컬럼을 기준으로 cc_df와 hour_cnt 데이터를 병합
```

```python
cc_df.head()
```

```python
cc_df.drop(['trans_date_trans_time', 'hour', 'hour_cat'], axis =1 , inplace = True)     # 시간 관련 feature 분석이 완료되었으므로 불필요해진 컬럼들을 제거
```

```python
# cc_num(신용카드 번호)를 기준으로 고객의 위치와 상점의 위치 간 거리의 z-score 계산하기
# 효과: 카드 사용자의 평소 거래 활동 반경을 벗어난 비정상 거래를 인식할 수 있습니다.
```

```python
from geopy.distance import distance
```

```python
cc_df['distance'] = cc_df.apply(lambda x: distance((x['lat'], x['long']), (x['merch_lat'], x['merch_long'])).km, axis = 1)      # 고객의 위치와 상점의 위치 사이의 거리를 계산하여 distance 컬럼에 저장
```

```python
from datetime import datetime
```

```python
cc_df.head()
```

```python
dist_info = cc_df.groupby('cc_num')['distance'].agg(['mean','std']).reset_index()       # cc_num 컬럼의 값마다 distance의 평균, 표준편차를 계산하여 dist_info 변수에 저장
```

```python
dist_info.to_pickle('./dist_info.pkl')
```

```python
cc_df = cc_df.merge(dist_info, on = 'cc_num', how = 'left')     # cc_num 컬럼을 기준으로 cc_df와 dist_info 데이터를 병합
```

```python
print(cc_df.shape)      # 행과 열의 개수 확인
```

```python
cc_df.head()
```

```python
cc_df['dist_z'] = (cc_df['distance'] - cc_df['mean']) / cc_df['std']        # distance 값의 z-score를 계산하여 dist_z 컬럼에 저장
```
